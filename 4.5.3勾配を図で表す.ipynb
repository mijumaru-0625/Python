```python
#  4-3-(1)
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#  -----
def f(w0, w1):                             # f 
    return w0 ** 2 + 2 * w0 * w1 + 3       #  4-41

def df_dw0(w0, w1):                        # f  w0 
    return 2 * w0 + 2 * w1                 #  4-44

def df_dw1(w0, w1):                        # f  w1 
    return 2 * w0                          #  4-46

#  -----
w0_min, w0_max = -2, 2
w1_min, w1_max = -2, 2
w0_n, w1_n = 17, 17
w0 = np.linspace(w0_min, w0_max, w0_n)
w1 = np.linspace(w1_min, w1_max, w1_n)
ww0, ww1 = np.meshgrid(w0, w1)           # 
f_num = f(ww0, ww1)                      # f 
df_dw0_num = df_dw0(ww0, ww1)            # f  w0 
df_dw1_num = df_dw1(ww0, ww1)            # f  w1 

#  -----
plt.figure(figsize=(9, 4))
plt.subplots_adjust(wspace=0.3)
# 
plt.subplot(1, 2, 1)
cont = plt.contour(                       # f 
    ww0, ww1, f_num, levels=10, colors="black")
cont.clabel(fmt="%d", fontsize=8)
plt.xticks(range(w0_min, w0_max + 1, 1))
plt.yticks(range(w1_min, w1_max + 1, 1))
plt.xlim(w0_min - 0.5, w0_max + 0.5)
plt.ylim(w1_min - 0.5, w1_max + 0.5)
plt.xlabel("$w_0$", fontsize=14)
plt.ylabel("$w_1$", fontsize=14)
# 
plt.subplot(1, 2, 2)
plt.quiver(                               # f 
    ww0, ww1, df_dw0_num, df_dw1_num)
plt.xlabel("$w_0$", fontsize=14)
plt.ylabel("$w_1$", fontsize=14)
plt.xticks(range(w0_min, w0_max + 1, 1))
plt.yticks(range(w1_min, w1_max + 1, 1))
plt.xlim(w0_min - 0.5, w0_max + 0.5)
plt.ylim(w1_min - 0.5, w1_max + 0.5)
plt.show()
```
